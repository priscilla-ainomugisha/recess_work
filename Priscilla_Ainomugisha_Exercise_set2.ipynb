# Series

Comparing Elements of two pandas series

```python
# A Pandas Series is a one-dimensional labeled array-like object provided 
# by the Pandas library in Python. 
# It is similar to a column in a spreadsheet or a one-dimensional array,
# but with additional functionalities and features.
################################################################
import pandas as pd

# Defining the two series
A = pd.Series([4, 65, 436, 3, 9])
B = pd.Series([7, 0, 3, 897, 9])

# Compare the elements of the two series
#This compares each individual element of the series
comparison = A == B

# Print the result
print(comparison)
```

Addition, subtraction, multiplication and Division of two series

```python
C = pd.Series([2, 4, 6, 8, 14])
D = pd.Series([1, 3, 5, 7, 9])
# Addition
addition = C + D
print("Addition:")
print(addition)

# Subtraction
subtraction = C- D
print("\nSubtraction:")
print(subtraction)

# Multiplication
multiplication = C * D
print("\nMultiplication:")
print(multiplication)

# Division
division = C / D
print("\nDivision:")
print(division)
```

Converting a dictionary to a series

```python

dictionary1 = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}

# Convert dictionary to series using pd.series
series = pd.Series(dictionary1)

print(series)
```

Converting a series to an array

```python
import numpy as np
# We use numpy to do this 

# Sample series
series = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])

# Convert series to array
array = np.array(series)

print(array)
```

Displaying most frequent value

```python
#reading the dataset
df=pd.read_csv('BSSE 2301 EXERCISE 2\AfricaCupofNationsMatches.csv')
```

```python
# Reading the column and assign it a vriable
serie_A = df['HomeTeamGoals']
```

```python
most_frequent_value = serie_A.value_counts().idxmax()#idxmax gets the index of the value most nö of goals

# Filtering out the most frequent goals, and replace the rest with other

serie_A = serie_A.apply(lambda x: most_frequent_value if x == most_frequent_value else 'Other')

print (serie_A)
```
