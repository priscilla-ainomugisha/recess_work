```python
# Data Analysis
# Checking for null values
df5.isnull().sum().sort_values(ascending = False)
```

```python
# Renaming the columns using another row
# Using DataFrame.rename() to convert row to column header
# This doens't remove the first row.
header_row = 1
df5.columns= df5.iloc[header_row]

print(df5.columns)
```

```python
df5.columns = df5.columns.str.replace(' ', '')
print(df5.columns)
```

```python
PERFORMANCE = df5[['CasessubmittedtoDPP/RSA','Caseswithconvictions','CaseswithAcquittals','CasesDismissed','CasespendinginCourt']].copy()
print(PERFORMANCE.to_string())
#TO HAVE THIS SCROLLABLE, YOU CLICK ON THE *SCROLLABLE ELEMENT* WITHOUT PRINTING ALL THE ROWS
```

```python
PERFORMANCE.isnull().sum().sort_values(ascending=False) 
```

```python
# Dropna() -Dropping rows with missing values
PERFORMANCE.dropna(inplace= True)
```

```python
m = PERFORMANCE['CasessubmittedtoDPP/RSA'].mean()
PERFORMANCE['CasessubmittedtoDPP/RSA'] = PERFORMANCE['CasessubmittedtoDPP/RSA'].fillna(m)

print(PERFORMANCE)
```

```python


# Convert the column to numeric values
PERFORMANCE['Cases with convictions'] = pd.to_numeric(PERFORMANCE['Cases with convictions'], errors='coerce')

# Verify the data type of the column
print(PERFORMANCE['Cases with convictions'].dtype)
```

```python
m = PERFORMANCE['Cases with convictions'].mean()
PERFORMANCE['Cases with convictions'] = PERFORMANCE['Cases with convictions'].fillna(m)
PERFORMANCE['Cases with convictions'] = PERFORMANCE['Cases with convictions'].fillna(m , inplace=True)
print(PERFORMANCE.to_string())
```
